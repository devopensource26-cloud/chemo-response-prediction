```python
from Bio import Entrez
import pandas as pd
import time

# Lista de genes = nombres de columnas, quitando las que no son genes
genes = [col for col in df.columns if col not in ["Response", "Unnamed: 0"]]

print(f"Genes detectados: {len(genes)}")

# Palabras clave optimizadas
cancer_terms = (
    '"breast cancer"[Title/Abstract] OR '
    '"mammary carcinoma"[Title/Abstract] OR '
    '"triple negative breast cancer"[Title/Abstract] OR '
    '"TNBC"[Title/Abstract]'
)
resistance_terms = (
    '"chemoresistance"[Title/Abstract] OR '
    '"drug resistance"[Title/Abstract] OR '
    '"chemotherapy resistance"[Title/Abstract] OR '
    '"chemotherapy response"[Title/Abstract] OR '
    '"treatment response"[Title/Abstract]'
)

# Función para contar publicaciones en PubMed
def count_pubmed_hits(gene, max_retries=3):
    query = f'"{gene}"[Title/Abstract] AND ({cancer_terms}) AND ({resistance_terms})'
    for attempt in range(max_retries):
        try:
            handle = Entrez.esearch(db="pubmed", term=query)
            record = Entrez.read(handle)
            handle.close()
            return int(record["Count"])
        except Exception as e:
            print(f"Error con {gene}, intento {attempt+1}: {e}")
            time.sleep(1)
    return None

# Crear lista de resultados
results = []
for i, gene in enumerate(genes, 1):
    hits = count_pubmed_hits(gene)
    results.append({"Gene": gene, "PubMed_hits": hits})
    print(f"{i}/{len(genes)} - {gene}: {hits}")
    time.sleep(0.4)  # respeta límite de peticiones

```

```python
# Guardar resultados en un DataFrame
out_df = pd.DataFrame(results)

# Exportar a CSV
out_df.to_csv("./data/processed/pubmed_gene_counts.csv", index=False)

print("✅ Resultados guardados en pubmed_gene_counts.csv")
```

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, GridSearchCV

# Cargar datos
X = df.drop(columns=["Response"])
y = df["Response"].astype(int).values

# Pipeline base
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(
        penalty="elasticnet",
        solver="saga",
        max_iter=2000,
        tol=1e-3,
        random_state=42,
        class_weight="balanced"
    ))
])

# Espacio de búsqueda
param_grid = {
    "clf__C": np.logspace(-3, 2, 20),         # 0.001 a 100 en escala log
    "clf__l1_ratio": [0.1, 0.3, 0.5, 0.7, 0.9]
}

# Validación cruzada estratificada
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Grid search
grid = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring="average_precision",   # AUPRC
    cv=cv,
    n_jobs=-1,
    verbose=2
)

# Entrenar búsqueda
grid.fit(X, y)

print("Mejor combinación encontrada:")
print(grid.best_params_)
print(f"Mejor AUPRC CV: {grid.best_score_:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.utils import resample
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import average_precision_score

# ======================
# Parámetros stability selection
# ======================
B = 50
sample_frac = 0.5
threshold = 0.45

# ======================
# Cargar datos de expresión
# ======================
X = df.drop(columns=["Response"])
y = df["Response"].astype(int).values

# ======================
# Cargar prioridades (solo log y binary)
# ======================
df_priority = pd.read_csv("./data/processed/genes_with_priority_variants.csv")
priority_types = ["priority_log", "priority_binary"]

priority_dicts = {}
for pt in priority_types:
    d = df_priority.set_index("Gene")[pt].to_dict()
    weights = pd.Series(X.columns.map(lambda g: d.get(g, 0.01)), index=X.columns)
    weights = weights.replace(0, 0.01)  # evitar ceros
    priority_dicts[pt] = weights

# ======================
# Hiperparámetros fijos (los mejores que ya tenías)
# ======================
best_params = {'C': 0.05623413251903491, 'l1_ratio': 0.9, 'class_weight': 'balanced'}

results = []

# ======================
# Stability selection con prioridades
# ======================
for pt in priority_types:
    print(f"\nProbando prioridad: {pt}")

    weights_raw = priority_dicts[pt]
    weights_alpha = weights_raw / weights_raw.max()

    # Aplicar pesos (genes con mayor prioridad quedan menos penalizados)
    X_weighted = X.copy()
    for gene in X.columns:
        w = weights_alpha.loc[gene]
        X_weighted[gene] = X[gene] / w

    # Pipeline
    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(
            penalty="elasticnet",
            solver="saga",
            max_iter=2000,
            tol=1e-3,
            random_state=42,
            n_jobs=-1,
            C=best_params['C'],
            l1_ratio=best_params['l1_ratio'],
            class_weight=best_params['class_weight']
        ))
    ])

    n_genes = X_weighted.shape[1]
    selection_freq = np.zeros(n_genes)

    # Stability selection
    for i in range(B):
        X_resampled, y_resampled = resample(
            X_weighted, y,
            replace=False,
            n_samples=int(sample_frac * len(y)),
            stratify=y,
            random_state=42 + i)

        pipe.fit(X_resampled, y_resampled)
        coefs = pipe.named_steps['clf'].coef_.ravel()
        selected = (np.abs(coefs) > 1e-8)
        selection_freq += selected.astype(int)

    selection_freq /= B

    stable_genes_idx = np.where(selection_freq >= threshold)[0]
    stable_genes = X_weighted.columns[stable_genes_idx]

    print(f"Genes estables (freq ≥ {threshold}): {len(stable_genes)}")

    # Evaluar AUPRC con genes estables
    if len(stable_genes) > 0:
        X_stable = X[stable_genes]

        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        auprc_scores = []
        for train_idx, test_idx in cv.split(X_stable, y):
            pipe.fit(X_stable.iloc[train_idx], y[train_idx])
            probs = pipe.predict_proba(X_stable.iloc[test_idx])[:, 1]
            score = average_precision_score(y[test_idx], probs)
            auprc_scores.append(score)
        mean_auprc = np.mean(auprc_scores)
    else:
        mean_auprc = np.nan

    print(f"AUPRC (CV) con genes estables: {mean_auprc:.4f}")

    results.append({
        'priority_type': pt,
        'n_stable_genes': len(stable_genes),
        'AUPRC_CV': mean_auprc,
        'stable_genes': stable_genes.tolist()
    })

# ======================
# Resumen final
# ======================
print("\n--- Resumen final ---")
for r in results:
    print(f"{r['priority_type']}: genes={r['n_stable_genes']}, AUPRC={r['AUPRC_CV']:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import roc_auc_score, make_scorer
from sklearn.preprocessing import StandardScaler
from stability_selection import StabilitySelection

# =======================
# 1. Cargar datos
# =======================
# df = tu matriz de expresión con "response"
df = pd.read_csv("./data/processed/matriz_genes.csv")  
y = df["response"].values
X = df.drop(columns=["response"])

# Prioridades de genes
priority_df = pd.read_csv("./data/processed/genes_with_priority_variants.csv")

# =======================
# 2. Mapear prioridades
# =======================
# Default prioridad = 1
priority_map = dict(zip(priority_df["Gene"], priority_df["priority_log"]))
priority_weights = np.array([priority_map.get(gene, 1.0) for gene in X.columns])

# Escalar X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =======================
# 3. Definir Stability Selection con Elastic Net
# =======================
def run_stability_selection(X, y, alpha, l1_ratio, random_state=42):
    base_model = LogisticRegression(
        penalty="elasticnet",
        solver="saga",
        l1_ratio=l1_ratio,
        C=1/alpha,   # inverso de alpha en sklearn
        max_iter=5000
    )
    
    stab_sel = StabilitySelection(
        base_estimator=base_model,
        n_jobs=-1,
        random_state=random_state,
        verbose=1
    )
    stab_sel.fit(X, y, sample_weights=priority_weights)  # aquí se usan tus prioridades
    
    selected = np.array(X.columns)[stab_sel.get_support()]
    return selected, stab_sel.scores_

# =======================
# 4. Búsqueda de hiperparámetros
# =======================
param_grid = {
    "alpha": [0.001, 0.01, 0.1, 1],
    "l1_ratio": [0.1, 0.5, 0.9]
}

best_score = -np.inf
best_params = None
best_genes = None

for alpha in param_grid["alpha"]:
    for l1 in param_grid["l1_ratio"]:
        print(f"Probando alpha={alpha}, l1_ratio={l1}")
        selected_genes, scores = run_stability_selection(X_scaled, y, alpha, l1)
        
        # Validación cruzada con los genes seleccionados
        if len(selected_genes) > 0:
            X_sel = X[selected_genes]
            cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
            aucs = []
            for train, test in cv.split(X_sel, y):
                model = LogisticRegression(penalty="elasticnet", solver="saga",
                                           l1_ratio=l1, C=1/alpha, max_iter=5000)
                model.fit(X_sel.iloc[train], y[train])
                y_pred = model.predict_proba(X_sel.iloc[test])[:, 1]
                aucs.append(roc_auc_score(y[test], y_pred))
            
            mean_auc = np.mean(aucs)
            print(f"AUC CV = {mean_auc:.3f} con {len(selected_genes)} genes")
            
            if mean_auc > best_score:
                best_score = mean_auc
                best_params = {"alpha": alpha, "l1_ratio": l1}
                best_genes = selected_genes

print("\nMejores parámetros encontrados:")
print(best_params)
print(f"Mejor AUC: {best_score:.3f}")
print(f"Genes seleccionados ({len(best_genes)}): {best_genes}")
```
